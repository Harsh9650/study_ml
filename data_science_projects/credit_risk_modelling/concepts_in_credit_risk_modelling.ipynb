# ðŸ’° Expected Loss (EL)

## ðŸ” Definition
Expected credit loss is the estimated amount a lender may lose by lending to a borrower.

## ðŸ’¡ Formula

$$
EL = PD \times LGD \times EAD
$$

**Where:**

- **EL** = Expected Loss  
- **PD** = Probability of Default  
- **LGD** = Loss Given Default  
- **EAD** = Exposure at Default  

---

## ðŸ“Š Component Derivation

### ðŸ”¹ Probability of Default (PD)

Suppose a lender has issued loans to 1,000 borrowers, and 25 defaulted over one year:

$$PD = \frac{\text{Number of Defaults}}{\text{Total Borrowers}} = \frac{25}{1000} = 0.025 \quad \text{(or 2.5\%)}$$

---

### ðŸ”¹ Loss Given Default (LGD)

If the average recovery is `$`60,000 from a `$`100,000 loan:

$$LGD = 1 - \frac{\text{Recovery}}{\text{Exposure}} = 1 - \frac{60,000}{100,000} = 0.40 \quad \text{(or 40\%)}$$

---

### ðŸ”¹ Exposure at Default (EAD)

If `$`80,000 is drawn from a `$`100,000 line of credit at the time of default:

$$
EAD = 80,000
$$

---

## ðŸ”— Putting It All Together

$$
EL = 0.025 \times 0.40 \times 80,000 = 800
$$

**Interpretation**: The lender expects to lose `$`800 per borrower on average based on historical data.

# ðŸ¦ Capital Adequacy and CAR

## ðŸ›¡ï¸ Definition

**Capital Adequacy** refers to a bankâ€™s ability to absorb losses through sufficient capital reserves. Frameworks like Basel II and III require banks to maintain a minimum ratio of capital relative to risk-weighted assets (RWA).

## ðŸ“ Capital Adequacy Ratio (CAR)

$$
CAR = \frac{\text{Capital}}{\text{Risk-Weighted Assets (RWA)}}
$$

**Capital** includes:
- **Tier 1 Capital**: Core capital (equity, reserves)
- **Tier 2 Capital**: Supplementary capital (subordinated debt, hybrid instruments)

---

### ðŸ”¢ Example

- Tier 1 Capital: `$`80 million  
- Tier 2 Capital: `$`20 million  
- RWA: `$`1,000 million

**Total Capital**:  
$$
80 + 20 = 100 \text{ million}
$$

**CAR**:  
$$
CAR = \frac{100}{1000} = 0.10 = 10\%
$$

**Interpretation**:  
Since the CAR is 10%, above the 8% regulatory minimum, the bank is well-capitalized.

# ðŸ›ï¸ Basel Accords: Basel I, II, III

## ðŸ”¹ Basel I (1988)

**Definition**:  
Focused primarily on **credit risk**, introducing the first international framework for **capital adequacy**.

**Key Features**:
- Minimum CAR of **8%** of risk-weighted assets
- Risk-weight buckets: 0%, 20%, 50%, 100%
- Only addressed **credit risk**

---

## ðŸ”¹ Basel II (2004)

**Definition**:  
Enhanced Basel I by introducing **market** and **operational risk**, and greater risk sensitivity.

**Three Pillars**:
1. **Minimum Capital Requirements**
2. **Supervisory Review Process**
3. **Market Discipline**

**Key Features**:
- Internal ratings-based approaches
- Operational risk added
- Emphasis on transparency and disclosure

---

## ðŸ”¹ Basel III (2010â€“2017)

**Definition**:  
Introduced after the 2008 financial crisis to improve shock absorption and reduce systemic risk.

**Key Features**:
- Increased minimum capital requirements
- Introduced **Capital Buffers** (Conservation and Countercyclical)
- Introduced **Leverage Ratio**
- Introduced **Liquidity Ratios**: LCR and NSFR
- Focused on **capital quality** (more CET1 capital)

---

## âœ… Summary

- **Basel I** laid the foundation for capital adequacy with a simple credit risk model.
- **Basel II** introduced more risks and better measurement methods.
- **Basel III** responded to financial crises by improving capital, liquidity, and leverage management.

## Credit Risk Approaches under Basel Framework

Banking regulators (under Basel II and III) require banks to calculate **credit risk** using one of the following three approaches:

### 1. Standardized Approach (SA)
- Uses regulator-defined risk weights.
- Relies on external credit ratings (e.g., Moodyâ€™s, S&P).
- Simpler and more conservative.
- Typically used by smaller or less complex banks.

### 2. Foundation Internal Ratings-Based Approach (FIRB)
- Banks estimate **Probability of Default (PD)** internally.
- **Loss Given Default (LGD)** and **Exposure at Default (EAD)** are provided by regulators.
- Requires supervisory approval and robust credit risk systems.

### 3. Advanced Internal Ratings-Based Approach (AIRB)
- Banks internally estimate all components: **PD**, **LGD**, and **EAD**.
- Offers more risk-sensitive capital requirements.
- Requires the highest level of internal risk management and regulatory oversight.

> ðŸ”’ Banks must get regulatory approval before using FIRB or AIRB.

# ðŸ“„ For Individual Borrowers Common Loan Types and Key Risk Factors

## 1. **Credit Card Loans**
- **Key Factors:**
  - Credit limit utilization (% of credit limit used)
  - Payment history (missed/delayed payments)
  - Number of open credit lines
  - Age of credit history
  - Recent inquiries

---

## 2. **Personal Loans (Unsecured)**
- **Key Factors:**
  - Debt-to-income ratio (DTI)
  - Credit score
  - Employment status and income stability
  - Loan amount vs income
  - Existing unsecured debt

---

## 3. **Auto Loans**
- **Key Factors:**
  - Loan-to-value (LTV) ratio of the vehicle
  - Vehicle age and depreciation
  - Borrower's income and employment status
  - Down payment made
  - Credit history

---

## 4. **Mortgage Loans (Home Loans)**
- **Key Factors:**
  - Loan-to-value (LTV) ratio
  - Debt-to-income (DTI) ratio
  - Credit score
  - Property location and type
  - Documentation type (full-doc, low-doc)

---


## 5. **Student Loans**
- **Key Factors:**
  - Educational institution type and reputation
  - Degree pursued (STEM vs non-STEM)
  - Repayment start date (grace period)
  - Borrower's cosigner status (if any)
  - Federal vs private lender (terms vary)

---

## ðŸ“Œ Note:

In credit risk modelling:

- **Probability of Default (PD)** is commonly estimated using **logistic regression**, since it is a binary outcome (default or no default).
- **Loss Given Default (LGD)** and **Exposure at Default (EAD)** are typically estimated using **regression techniques**, such as **beta regression**, **linear regression**, or **Tobit models**, depending on the data characteristics (e.g., bounded outcomes for LGD).

> âœ… Logistic regression for PD is well-established.
> âœ… LGD and EAD require continuous outcome models, beta regression is used especially when outcomes are proportions (0 to 1).
